Recursive function: A function that calls itself.

```python
def factorial(n):
    if(n==0):
        return 1
    else:
        
        return n* factorial(n-1)
    
```

```python
factorial(6)
```

```python
def fibonacci(n):
    if(n==0 or n==1):
        return n
    return fibonacci(n-1)+fibonacci(n-2)
```

```python
fibonacci(4)
```

Decorator function:

```python
def div(a,b):
    c=a/b
    return c
```

```python
div(10,5)
```

```python
## will swap the numbers if a is less than b
def div(a,b):
    if a<b:
        a,b = b,a
    c=a/b
    return c
```

```python
div(10,5)
```

```python
div(2,10)
```

```python
def og_div(a,b):
    c=a/b
    return c
```

```python
def smart_div(func):
    def inner(a,b):
        if a<b:
            a,b=b,a
        return func(a,b)
    return inner
```

```python
result=smart_div(og_div)
```

```python
result(5,10)
```

```python
result=smart_div(og_div(5,10))
result
```

Task: Logging Decorator

 

You are tasked with creating a decorator function called log_function_call that logs when a function is called, along with its arguments, and the result of the function call. You will apply this decorator to two functions.

 

Implement a decorator function called log_function_call that takes a function as an argument and wraps it. Inside the decorator, log the function's name, its arguments, and the result of the function call. You can use the logging module for logging.

 

Create two functions:

 

add(a, b): This function takes two integers, a and b, as arguments and returns their sum.
multiply(a, b): This function takes two integers, a and b, as arguments and returns their product.
Apply the log_function_call decorator to both the add and multiply functions.

 

Call the decorated add and multiply functions with different arguments to verify that the decorator logs the function calls and results correctly.

```python
def add(a,b):
    return a+b
```

```python
def multiply(a,b):
    return a*b
```

```python
def log_function_call(func):
    def inner(a,b):
        func(a,b)
    return inner
res=log_function_call(add)
res2= log_function_call(multiply)
```

```python
res(8,9)
res
```

```python
res2(4,7)
res2
```

Generator function: Yield keyword to yield a sequence of values one at a time, instead of returning all at once.

```python
def count_up_to(n):
    i=1
    while(i<=n):
        yield 1
        i=i+1
```

```python
count_up_to(10)
```

```python
for i in count_up_to(10):
    print(i)
```

Filter, map , lambda functions are important for data transformations.

```python
def evenodd(x):
    if x%2==0:
        return True
    else:
        return False
    
```

```python
evenodd(6)
```

```python
l=[0,2,3,4,5,6,8]
li=list(filter(evenodd,l))
li
```

```python
li=list(filter(lambda x: x%2==0,l))
li
```

```python
l=[10,11,12,14,15]
```

```python
def squareit(n):
    return n**2
```

```python
sq= list(map(squareit,l))
sq
```
