Create a list of months: ['January', 'February', 'March', 'April', 'May'].
Create a list of monthly sales data for a hypothetical product: [2500, 3200, 2800, 4100, 3700]

```python
months= ['January', 'February', 'March', 'April', 'May']
sales_data= [2500, 3200, 2800, 4100, 3700]
```

```python
months= np.array(months)
sales_data= np.array(sales_data)
```

```python
plt.bar(months, sales_data)
plt.title("Monthly Sales Data")
plt.xlabel("months")
plt.ylabel("sales")
```
